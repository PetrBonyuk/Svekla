Наверное мы должны были подумать, что она действительно испытывает к нему какие-то чувства. Однако она довольно просто соглашается на предательство.

Перед битвой на Зелёном Зубце Тирион просит найти ему девушку:

> Лучше, чтоб была помоложе и посимпатичнее, но – какую найдешь. Я уже буду рад, если она хоть раз умывалась в этом году. Если нет, пусть умоется. Но скажи ей, кто я, и предупреди, каков из себя.

> **Тирион (Тирион VIII, Игра Престолов)**:

Бронн приводит Шаю. И вот что странно, Шая - слишком привлекательная. Настолько, что Тирион предпочитает её даже в сравнении с девушками из борделя Катайи, в ведь это - самый элитный бордель Королевской гавани, который могут позволить себе только лорды.

> Подняв свечу, он оглядел ее. Бронн постарался: тоненькая, с глазами голубки, твердые и маленькие грудки… и улыбка, то застенчивая, то наглая и ехидная. Ему это понравилось.
> **(Тирион VIII, Игра Престолов)**

И Серсея именно её находит привлекателькой в сравении с другими благородными дамами при дворе:

> Хорошеньких, вроде горничной леди Лоллис, ждет горячая ночка, но старым, большим и безобразным тоже нечего ждать пощады.
> **Серсея (Санса VI, Битва Королей)**

Санса тоже это подмечает:

> Горничной Лоллис, хорошенькой девушке с короткими темными волосами, похоже, как раз очень хотелось спихнуть свою госпожу в сухой ров, прямо на железные пики.
> **(Санса V, Битва Королей)**

Тирион спрашивает как он умудрился её найти:

> Отобрал у рыцаря. Он не хотел расставаться с ней, но ваше имя несколько переменило его настроение… ну и мой кинжал у горла.

Это странно. Если он отпустил Шаю от имени Тириона Ланнистера, то зачем нож у горла? Кто поверит Бронну на слово? Он, вероятно, отобрал его у какого-то рыцаря, и если дело в кинжале у горла, то, значит, между ними должна была быть драка. И разве в таком случае она не должна быть очень напугана?

> Кажется, я просил найти мне шлюху, а не врага.
> **Тирион (Тирион VIII, Игра Престолов)**

Не говоря уже о том, что Бронн превысил свои полномочия. Если бы не нашел какой-нибудь свободной симпатичной девушки, он должен был подойти к Тириону и попросить золота, чтобы заплатить за занятую девушку.

Тирион спросил, у кого отобрал ее Бронн, Шая назвала мелкого вассала кого-то из незначительных лордов.

> Его можно не опасаться, милорд, он не из больших людей.
> **Шая (Тирион VIII, Игра Престолов)**

Удивительно, что настолько молодая и привлекательная проститутка была с очевидно не самым богатым человеком. И это ещё раз подчёркивает неадекватность действий Бронна. Если он был мелким рыцарем, ему можно было относительно немного заплатить. Это особенно странно для человека, который был особенно дальновидным в таких ситуациях.

## Связь с Варисом

По прибытии в Королевскую гавань Тирион оставляет шаю в гостинице под охраной. Однако Варис моментально узнаёт об этом и после посещения малого совета они встречаются в этой гостинице.

> Лорд Варис? – опешил Тирион. – Не ожидал встретить вас здесь. – Иные его побери: как он ухитрился разыскать их так быстро?
> **(Тирион I, Битва Королей)**

Варис объясняет это так:

> **(Тирион I, Битва Королей)**

> Я не заметила, милорд. Завтра схожу посмотрю, если хотите.
> **Шая**

> «Не беспокойся, милая, – подумал Тирион, покачивая свой кубок. – Плевать он хотел на эти фигуры. Глаза, о которых он толкует, – его собственные. Он хочет сказать, что следил за нами, и узнал, что мы здесь, в тот же миг, как мы проехали в ворота».
> **(Тирион I, Битва Королей)**

Да, нетрудно было вычислить, что Тирион разместил в гостиннице кого-то, однако он не мог знать про любовницу. И простая слежка этого не объясняет. Тирион знал о репутации Вариса, как в общем то и все:

> Тыше ты, дурень, — забеспокоился Армин. — Никогда не шути такими вещами. Откуда тебе знать, кто тебя слышит. У Паука везде уши.
> **(Пролог, Пир Воронов)**

> Квентин уже начал думать, что лучше бы им было купить собственный корабль в Дощатом городе. Но это привлекло бы к ним ненужное внимание: у Паука везде были глаза и уши, даже в стенах Солнечного Копья.

Даже в Староместе и Солнечном Копье люди опасались говорить что-то лишнее. И Тирион в Королевской гавани должен быть в разы осторожнее. А раз он действительно был удивлён тем, что Варис нашёл Шаю, он действовал максимально скрытно.

Теперь, чтобы Тирион мог встретится с Шаей, ему необходимо обратится к Варису. В добавок даёт Варису кредит доверия, ведь он не рассказывает Серсее о Шае.

В наши смутные времена повсюду подстерегают злодеи, да-да. Люди с холодной сталью и еще более холодными сердцами. – «Куда я пришел один и без оружия, могут прийти и другие с мечами в руках», – подразумевала его речь.

> С вашей молодой леди очень приятно беседовать. На вашем месте я очень хорошо заботился бы о ней.

### Речь на суде

> Кое-что я подслушала, милорд, кое о чем миледи сама проговорилась. Но почти все остальное я слышала от него самого. ### Другие странности

У потенциального романа Шаи и Тайвина есть множество других проблем.

1. **Почему Шая?**
Тайвин - один из самых богатых людей Вестероса, если не самый богатый. У него есть выбор из тысяч женщин. И из всех он выбирает Шаю? Из всех женщин в мире - девушка своего сына? К тому же она уже показала, что не из молчаливых, а для Тайвина это важно.

3. **Как она там оказалась?**
Очевидно, что Тайвин не позволил бы, известной на весь замок шалаве проходить через обычную дверь. На самом деле, единственный вариант, как она там оказалась - потайные проходы. Ведь она уже проникала в башню десницы:

> На кровати под балдахином, поджав ноги, сидела Шая – голая, но с золотой цепью на груди. Цепь была составлена из золотых рук, сжимающих одна другую.
> **(Тирион XII, Битва Королей)**

> Она села, и одеяло соскользнуло с нее. Под ним она была голая, если не считать цепи на шее. Цепь состояла из золотых рук, каждая из которых сжимал запястье другой.
> **(Тирион XI, Буря Мечей)**

Эти сцены явно параллельны. И Шая рассказывает, как проникла в первый раз:

> Лорд Варис завязал мне глаза, и я ничего не видела. Только в одном месте разглядела пол… там такая картинка выложена из мелких кусочков.
> **(Тирион XII, Битва Королей)**

Так что скорее всего Шая оказалась там, потому что Варис её туда провёл.

4. **Состояние Тайвина**
В последних главах его неоднократно описывали, как отравленного. И существует большая вероятность, что его отравили Мартеллы. Судя по болезненнуму виду, он был не в лучших кондициях, чтобы заниматься сексом.

Противоречий слишком много, чтобы игнорировать их. Поэтому мы должны прийти к очевидному выводу. Шая не спала с Тайвином. Она оказалась в его постели незадолго до того, как туда пришел Тирион. Варис избавился от неё, потому что она перестала быть полезной, а также потому что много знала.  

## Почему она делала то, что делала?

Если она агент Вариса, довольно трудно определить её мотивацию. Дело явно не в золоте, иначе под страхом смерти она выдала бы Вариса. Значит он повязал её чем-то сильнее. Возможно взял в заложники кого-то из её родственников, и я бы обратил внимание на эту фразу:

> Я могла бы родить тебе сыновей, знаю, что могла бы…
> **Шая, (Тирион X, Битва Королей)**

Возможно у Шаи есть ребёнок, которого Варис держит в заложниках. И это бы объяснило всё. А также сделало бы историю Шаи куда более трагичной. Это делает поступок Тириона ещё зловещее, а так же показывает, насколько Варис - гандон и манипулятор. Понятно, почему Мартин назвал убийство Шаи - худшим поступком Тириона.

# Бронн

Если Шая - агент Вариса, то и Бронн тоже, ведь именно он познакомил её с Тирионом. Но есть ли другие причины считать так?

## Действия в Пире

Бронн преподносится как умный персонаж. Но в Пире воронов он творит полное безумие. Он узурпирует Стокворт, зная, что это спровоцирует ответные действия Королевской гавани. И в условиях, когда Ланнистеры победили - это наиглупейший поступок. Конечно если не знаешь про дело Эйгона и вторжение Золотой компании.

И если мы это примим, многое встанет на свои места.

### Гостиница на перекрестке

Варис знал, что Кейтилин Старк идёт на север с намереньем поймать Тириона. Поэтому с его стороны было бы разумно отправить своего агента рядом с ней, на случай встречи Кейтилин и Тириона. Бронн - один из лучших бойцов Семи корлевств, поэтому его присутствие в Гостинице в роли межевого рыцаря более чем подозрительно.

Другой вариант: гостиница на перекрестке - база для людей Вариса. Ей управляет дом Хедль. Во время второго восстания Блэкфайера дом Хедль - был одним из самых главных заговорщиков. И сейчас они поддерживают Братство без знамён, которое настроено против Ланнистеров. И именно у Перекрёстка встретились войска Тайвина и Тирион, где Бронн смог бы прокординировать действия с Шаей.

### Как Варис вычислил Шаю?

В первой главе Тирон оставляет Бронна и Тиммета возле двери в зал Малого совета:

> Это Тиметт, сын Тиметта, Красная Рука клана Обгорелых. А это Бронн.
> ...
> Вы можете пройти, но без них.
> **(Тирион I, Битва Королей)**

А когда Тирион вернулся его ждал, только Бронн:

> Выйдя из зала совета, Тирион кивнул сиру Мендону и зашагал прочь по длинному сводчатому коридору. Бронн поравнялся с ним, но Тиметт, сын Тиметта, исчез бесследно.
>
> – А где наша Красная Рука? – осведомился Тирион.
>
> – Пошел осматривать замок. Такие, как он, не созданы для долгого ожидания за дверью.
> **(Тирион I, Битва Королей)**

И интересно, что Варис ушел первым:

> Варис встал первым, улыбаясь умильно, как всегда.
> **(Тирион I, Битва Королей)**

Таким образом, Бронн отпустил Тиммета, чтобы проинформировать Вариса о положении вещей.
